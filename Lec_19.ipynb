```python
number= int(input("please enter a number: "))

if number == 0:
  print("even")
elif number%2 ==0:
  print("even")
else:
  print("odd")
```

```python
number= int(input("please enter a number: "))

if number%10 ==0 and number%5 == 0:
  print('Number ends with 0')
elif number%5 == 0:
  print("Number ends with 5")
else:
  print("Number ends with some other number")
```

```python
marks = int(input("please enter your marks: "))

if 100>= marks >=0:
  if marks>= 90:
    print('Grade: A')
  elif marks>= 80:
    print('Grade: B')
  elif marks>=70:
    print('Grade: C')
  elif marks >= 60:
    print('Grade: D')
  else: 
    print('Grade: E')
else:
  print('Invalid Input')
```

```python
time = str(input('Travel from city A to city B: longer or shorter')).lower()

if time == 'longer':
  price = str(input('Price: higher or lower')).lower()
  if price == 'lower':
    print('Coach')
  if price == 'higher':
    print('Train')

if time == 'shorter':
  price = str(input('Price: higher or lower')).lower()
  if price == 'lower':
    print('Red Eye Flight')
  if price == 'higher':
    print('Daytime flight')

print("Arrive city B")


```
