different ways of writing print statements

```python
num = int(input())

for i in range(1,11):
  print(num, 'X',i, '=', i*num)
  print(f'{num} X {i} = {num*i}')   #formatted string
  print('{0} X {1} = {2}'.format(num, i, num*i))
  print('%d X %d = %d' % (num,i,num*i))  #using modulo operator
```

```python
pi = 22/7

print(f'Value of pi = {pi:.2f}')
```
