Dynamic Typing

```python
a = 10 
print(type(a))
```

Python allows us to use a variable to store a int as well as str later


```python
n = 10     #integer 
print(n)
print(type(n))

n = n/2   # becomes float the moment you divide
print(n)

print(type(n))

c=6

c = c/1
print(c)

print(type(c))
```
