Type Casting : Conversion of one data type into another. eg: from int to str, str to float, etc

```python
a = int(5.7)
b = int('5')
print(a, type(a))
print(b, type(b))
c = int(10.5)
print(c)
```

```python
a = float(9)
b= float('5.3')
print(a, type(a))
print(b,type(b))
```

```python
a = str(9)
b= str('5.3')
print(a, type(a))
print(b,type(b))
```

```python
a = bool(10)
b= bool(0)
c= bool(-10)

print(a, type(a))
print(b,type(b))
print(c, type(c))

```

```python
a = bool(10.5)
b= bool(0)
c= bool(-10.4)

print(a, type(a))
print(b,type(b))
print(c, type(c))
```
