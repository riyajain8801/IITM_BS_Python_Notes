```python
def list_min (l):
  mini = l[0]
  for i in range(len(l)):
    if (l[i]< mini) :
      mini = l[i]
  return mini
```

```python
l = [7,1,-5]
list_min(l)
```

```python
def list_max (l):
  maxi = l[0]
  for i in range(len(l)):
    if (l[i] > maxi):
      maxi = l[i]
  return maxi
```

```python
l = [10,23,35]

list_max(l)
```

```python
def list_appendbefore (l,z):
  newl = l + z
  return newl

```

```python
l1 =[10,20]
l2 = [2,3]

list_appendbefore(l1,l2)

```
