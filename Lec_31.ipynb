```python
print("Welcome to Trade Calc")

p = 0   # total profit

print("Please enter your emp ID:", end='')
n = input()
print(n)

while n != '-1':

    print('Please enter your trade amount (0 to stop): ', end='')
    t = int(input())
    print(t)

    while t != 0:
        p = p + t
        print('Enter next trade amount: ', end='')
        t = int(input())
        print(t)

    print("Total Profit or Loss is:", p)

    print("\nPlease enter your emp ID:", end='')
    n = input()
    print(n)

print("Thank you!")



```
