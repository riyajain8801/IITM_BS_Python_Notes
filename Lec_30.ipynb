```python
#factorial of a given number using for loop

n = int(input("please enter a number: "))

print(n)


factorial = 1


for i in range(1, n+1):
  factorial = factorial*i


print(factorial)


```

```python
#calculating the length of the number

n = str(input("Please enter a number: "))

count = 0

for i in n:
  count = count+1

print(count)
```

```python
# Reversing the number of digits


n = str(input("Please enter a number: "))

print(n)

rev = ''

for i in n:
  rev = i + rev

print(rev)
```
